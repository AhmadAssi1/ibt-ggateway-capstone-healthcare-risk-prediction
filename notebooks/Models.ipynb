```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, average_precision_score
path = "/data/"
X_train=pd.read_csv(path + "X_train.csv")
X_test=pd.read_csv(path + "X_test.csv")
y_train=pd.read_csv(path + "y_train.csv")
y_test=pd.read_csv(path + "y_test.csv")
results={}
y_train = y_train.values.ravel()
y_test = y_test.values.ravel()
def LogReg(X_train,X_test,y_train,y_test):
   #transform from DF into vector

    pipeline =Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(max_iter=1000,class_weight="balanced",random_state=42))
])

    pipeline.fit(X_train,y_train)

    y_pred = pipeline.predict(X_test)
    y_prob = pipeline.predict_proba(X_test)[:,1]

    # Predict probabilities (for AUROC / AUPRC)
    if hasattr(pipeline, "predict_proba"):
        y_prob = pipeline.predict_proba(X_test)[:, 1]  # probability of class 1
    else:
        y_prob = y_pred  # fallback (e.g., for regressors)

    # Metrics
    metrics = {
        "auroc": roc_auc_score(y_test, y_prob),
        "auprc": average_precision_score(y_test, y_prob)
    }

    for k,v in metrics.items():
         print(f" [LogisticRegression] {k}: {v:.4f}")
    results["LogReg"] = metrics
LogReg(X_train, X_test, y_train, y_test )
```
